```python
def tran_ru_en(t):
  result = [translator.translate(t1,src='ru', dest='en').text for t1 in t]
  return result
```

```python
def tran_en_ru(t):
  result = [translator.translate(t1,src='en', dest='ru').text for t1 in t]
  return result
```

```python
translated = df.apply(tran_en_ru,axis=1,result_type='expand')
```
